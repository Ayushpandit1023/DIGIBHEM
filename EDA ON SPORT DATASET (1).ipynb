# Visualize

```python
# season in which most matches were played
plt.figure(figsize = (10,5))
sns.countplot(x = 'season',data=matches, palette = 'Accent')

plt.xlabel('seasons')
plt.ylabel('No. of matches')
plt.show()
```

```python
# cities which host most No. of matches
plt.figure(figsize = (10,5))

sns.countplot(x = 'city',data=matches, palette = 'Accent')
plt.xlabel('Cities')
plt.ylabel('No. of matches')
plt.xticks(rotation=90)
plt.show()
```

```python
# Stadium which host most No. of matches
plt.figure(figsize = (10,5))

sns.countplot(x = 'venue',data=matches, palette = 'Greens_r')
plt.xlabel('Stadium')
plt.ylabel('No. of matches')
plt.xticks(rotation=90)
plt.show()
```

# toss decisions

```python
field_first = matches['toss_decision'].value_counts()[0]
bat_first = matches['toss_decision'].value_counts()[1]
```

```python
category_names = ['Ball First','Bat First']
sizes = [field_first, bat_first]
custom_colours = ['#ff7898','#74b9ff']

plt.figure(figsize=(2,2), dpi=226)
plt.pie(sizes, labels=category_names,
        textprops={'fontsize': 6},
       autopct="%1.2f%%", colors = custom_colours,
       explode=[0,0.05])
plt.show()

       
```

# toss decision seasonwise

```python
plt.subplots(figsize=(10,6))
sns.countplot(x='season',hue='toss_decision',data = matches)
plt.show()
```

# max toss winnner team

```python
toss_winner1=matches['toss_winner'].unique()
toss_winner2=matches['toss_winner'].value_counts().sort_index()

plt.figure(figsize=(10,5))
plt.title('Toss Winner',fontsize=13)
plt.xlabel('team Names')
plt.ylabel('no.of wins')
plt.xticks(rotation=90)

sns.barplot(x = toss_winner1,y = toss_winner2, palette='RdYlGn')
plt.show()
```
